## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
pymac_complete = pd.merge(mouse_metadata, study_results, how="outer", on= "Mouse ID")

# Display the data table for preview
pymac_complete
```

```python
# Checking the number of mice.
mice_count = pymac_complete["Mouse ID"].value_counts()
mice_count_total = len(mice_count)
mice_count_total
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = pymac_complete.loc[pymac_complete.duplicated
                                    (subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_ID = pymac_complete[pymac_complete['Mouse ID'].isin(duplicate_mice)==True]
duplicate_mice_ID
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_df = pymac_complete[pymac_complete['Mouse ID'].isin(duplicate_mice)==False]
new_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice=new_df["Mouse ID"].value_counts()
clean_number_of_mice=len(clean_mice)
clean_number_of_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
regimen_mean = new_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
regimen_mean

regimen_med = new_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
regimen_med

regimen_var = new_df.groupby('Drug Regimen').var()["Tumor Volume (mm3)"]
regimen_var

regimen_std = new_df.groupby('Drug Regimen').std()["Tumor Volume (mm3)"]
regimen_std

regimen_sem = new_df.groupby('Drug Regimen').sem()["Tumor Volume (mm3)"]
regimen_sem

# Assemble the resulting series into a single summary dataframe.
summary_df = pd.DataFrame({"Mean": regimen_mean, "Median": regimen_med,
                            "Variance":regimen_var, "Standard Deviation": regimen_std,
                           "SEM": regimen_sem})

summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
single_line_groupby = new_df.groupby("Drug Regimen")
summary_agg = single_line_groupby.agg(["mean","median","var","std","sem"])["Tumor Volume (mm3)"]
summary_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
regimen_df = new_df.loc[(new_df["Drug Regimen"] == "Capomulin") | 
                        (new_df["Drug Regimen"] == "Ramicane") |
                        (new_df["Drug Regimen"] == "Infubinol") | 
                        (new_df["Drug Regimen"] == "Ceftamin")] 
                        

regimen_df

# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

reg_max = regimen_df.groupby("Mouse ID").max()["Timepoint"]
reg_max

reg_df = pd.DataFrame(reg_max)
reg_merge = pd.merge(reg_df, new_df, on=("Mouse ID","Timepoint"), how ="left")
reg_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume = []

# Create for loop to loop through each drug
for drug in drugs:
    
    temp_df = reg_merge.loc[reg_merge['Drug Regimen'] == drug]

    # Create subset based on final Timepoint
    final_volume_df = temp_df.loc[temp_df['Timepoint'] == temp_df['Timepoint']]
    
    # Create series for final volume and append to empty list
    values = final_volume_df['Tumor Volume (mm3)']
    tumor_volume.append(values)
    
    # Calculate IQR
    quartiles = values.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    # Find upper/lower bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)

    # Check for outliers
    outliers_count = (values.loc[(final_volume_df['Tumor Volume (mm3)'] > upper_bound) | 
                                        (final_volume_df['Tumor Volume (mm3)'] < lower_bound)]).count()

    # Print summary
    print(f'The IQR for {drug} is {iqr}')
    print(f'The Lower Bound for {drug} is {lower_bound}')
    print(f'The Upper Bound for {drug} is {upper_bound}')
    print(f'Number of {drug} outliers is {outliers_count}')
  
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
flierprops = dict(marker='o', markerfacecolor='r', markersize=10, markeredgecolor='b')

# Plot boxplot
plt.boxplot(tumor_volume, labels = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'],  flierprops=flierprops)

# Add title and labels to boxplot
plt.title('Final Tumor Volume by Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Identify a mouse treated with Capomulin from df without duplicates
capomulin_mouse = new_df.loc[new_df['Mouse ID'] == 'y793']

plt.plot(capomulin_mouse['Timepoint'], capomulin_mouse['Tumor Volume (mm3)'], marker = 'o')


plt.xlabel("Time")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment of Mouse y793")

plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
# Create df for Capomulin drug regimen ONLY. 
cap_reg = new_df.loc[new_df["Drug Regimen"] == "Capomulin"]
cap_reg

# Find the average tumor volume for every mouse treated with Capomulin

avg_tumor_vol = cap_reg.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Find the average weight for every mouse treated with Capomulin
avg_weight = cap_reg.groupby('Mouse ID')['Weight (g)'].mean()

x = avg_weight
y = avg_tumor_vol

plt.scatter(x,y)

plt.xlabel("Weight")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Average Tumor Volume by Weight')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Calculate correlation from previous scatterplot
correlation = st.pearsonr(x,y)

# Calculate linear regression 
# from scipy.stats import linregress in cell 1
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress_values = x * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x,y)
plt.plot(x,regress_values,"r-")

# Set labels and plot
plt.xlabel("Weight")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Average Tumor Volume by Weight')
plt.annotate(line_eq, (20, 38), fontsize=16, color="red")
plt.show()

# Print correlation efficient
print(f"""The correlation between weight and average tumor volume
on the Capomulin regimen is {round(correlation[0],4)}.""")
```
